```python
# Import the necessary libraries for reading csv files
import os
import csv
from pathlib import Path

# Set the path for the csv file
csvpath = Path("Resources/budget_data.csv")

#Create Variables to Store the data
Total_Months = 0
Total = 0
Average_Change = 0
Last_Profit = 0
New_Profit = 0
Change = 0
Total_Change = 0
Greatest_Increase_Profits = 0
Greatest_Decrease_Profits = 0
Date_Increase_Profits = ""
Date_Decrease_Profits = ""


# Open the csv
with open(csvpath, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")

# Iterate through the data and skip the header of the CSV.
    csv_header = next(csvreader)

#* The total number of months included in the dataset.

    for row in csvreader:
  
        if Total_Months == 0:
            New_Profit = int(row[1])
        #Change += (New_Profit - Last_Profit)
            Last_Profit = New_Profit
        
        else:
            New_Profit = int(row[1])
            Total_Change += (New_Profit - Last_Profit)
            Change = (New_Profit - Last_Profit)
            Last_Profit = New_Profit
            
            # Conditional formulas to calculate the Greatest Increase and Decrease in profits
            if Change > Greatest_Increase_Profits:
                Greatest_Increase_Profits = Change
                Date_Increase_Profits = str(row[0])
            
            elif Change < Greatest_Decrease_Profits:
                Greatest_Decrease_Profits = Change
                Date_Decrease_Profits = str(row[0])
            
        # Keep a running total of the months and the Total Profit/Loss
        Total_Months += 1
        Total += int(row[1])
        

# Calculate the Average Change
Average_Change = round(Total_Change/(Total_Months-1),2)        

# Print the Results
print("Financial Analysis")
print ("------------------")
print(F"Total Months: {Total_Months}")
print(F"Total: ${Total}")
print(F"Average_Change: ${Average_Change}")
print (f"Greatest Increase in Profits: {Date_Increase_Profits} (${Greatest_Increase_Profits})")
print (f"Greatest Decrease in Profits: {Date_Decrease_Profits} (${Greatest_Decrease_Profits})")

output_path = Path("output.txt")

# Open the output_path as a file object in "write" mode ('w')

with open(output_path, 'w') as file:
    file.write("Financial Analysis\n")
    file.write ("------------------\n")
    file.write (F"Total Months: {Total_Months}\n")
    file.write (F"Total: ${Total}\n")
    file.write (F"Average_Change: ${Average_Change}\n")
    file.write (f"Greatest Increase in Profits: {Date_Increase_Profits} (${Greatest_Increase_Profits})\n")
    file.write (f"Greatest Decrease in Profits: {Date_Decrease_Profits} (${Greatest_Decrease_Profits})\n")



```
