```python
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path("Resources/menu_data.csv")
sales_filepath = Path("Resources/sales_data.csv")

# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []

# @TODO: Read in the menu data into the menu list

# Open the csv
with open(menu_filepath, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")

# Iterate through the data and skip the header of the CSV.
    csv_header = next(csvreader)

    for row in csvreader:
        menu.append(row)
        
# @TODO: Read in the sales data into the sales list

with open(sales_filepath, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")

# Iterate through the data and skip the header of the CSV.
    csv_header = next(csvreader)

    for row in csvreader:
        sales.append(row)

# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}

# Initialize a row counter variable
row_count = 0

# @TODO: Loop over every row in the sales list object
       
# @TODO: Increment the row counter by 1


# @TODO: Print total number of records in sales data


sales_quantity = {}

for row in sales:
    
    row_count +=1
    if row[4] not in sales_quantity:
        sales_quantity.update({row[4]: int(row[3])})
    
    elif row[4] in sales_quantity:
        key = str(row[4])
        sales_quantity[key] +=  int(row[3])

print(row_count)        
    
for key, value in sales_quantity.items():
    if key not in report:
        report[key] = {
    "01-count": value,
    "02-revenue": 0,
    "03-cogs": 0,
    "04-profit": 0,
    }


menu_price_cost ={}
price_cost_list = []

for row in menu:
    
    if row[0] not in menu_price_cost:
        menu_price_cost.update({row[0]: [float(row[3]),float(row[4])]})

# @TODO: Print out matching menu data
# @TODO: Else, the sales item does not equal any fo the item in the menu data, therefore no match
        
for key,value in sales_quantity.items():
    if key in menu_price_cost:
        print(key)
        price_cost_list = menu_price_cost[key]
        price_cost_list.append(value)
        price_cost_list.append(price_cost_list[0]-price_cost_list[1])
        menu_price_cost[key] = price_cost_list
    else:
        print("Not matching menu data")

 
for key, value in menu_price_cost.items():
    if key in report:
        report[key] = {
        "01-count": value[2],
        "02-revenue": value[0]*value[2],
        "03-cogs": value[1]*value[2],
        "04-profit": value[3]*value[2],
        }


 
# @TODO: Write out report to a text file (won't appear on the command line output)

output_path = Path("output.txt")
with open(output_path, 'w') as file:
    text = str(report)
    file.write(text)




```
